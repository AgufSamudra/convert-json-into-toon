<br>

Penggunaan TOON format terbukti jauh lebih efisien dibandingkan dengan format JSON. TOON format secara konsisten menghasilkan pengurangan jumlah token yang signifikan. Hal ini berdampak langsung pada penghematan biaya penggunaan LLM, karena biaya dihitung berdasarkan jumlah token input dan output. Dengan demikian, TOON format dapat menjadi strategi efektif untuk mengoptimalkan performa sekaligus menekan biaya operasional model bahasa besar (LLM).
